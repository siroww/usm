## 🧩 Модель данных

```python
from pydantic import BaseModel, Field
from typing import Optional
from datetime import datetime

class Task(BaseModel):
    title: str = Field(..., min_length=1, max_length=100)
    description: Optional[str] = None
    priority: str = Field(default="medium", pattern="^(low|medium|high)$")
    completed: bool = False
    deadline: Optional[datetime] = None
```

👉 При возврате задачи API должен добавлять:

* `id`
* `created_at`
* `completed_at` (если выполнена)


## 🧩 Часть 1: Базовые эндпоинты (30 баллов)

### 1. GET /

Возвращает:

```json
{
  "message": "Advanced Task API",
  "version": "2.0"
}
```


### 2. GET /tasks

Поддерживает параметры:

* `completed: bool`
* `priority: str`
* `limit: int`
* `sort_by: created_at | priority`

📌 Логика:

* Фильтрация по параметрам
* Сортировка:

  * priority: high → medium → low
* Ограничение количества


### 3. GET /tasks/{task_id}

* Возвращает задачу
* Ошибка 404 если не существует


### 4. POST /tasks

📌 Дополнительно:

* Проверка уникальности `title`
* Автоматически:

  * `id`
  * `created_at`

📌 Ошибки:

* 400 если title уже существует

## 🧩 Часть 2: Расширенные операции (40 баллов)


### 1. PUT /tasks/{task_id}

Полное обновление:

* Все поля обязательны
* Нельзя менять:

  * `id`
  * `created_at`


### 2. PATCH /tasks/{task_id}

Частичное обновление:

* Можно менять:

  * title
  * description
  * priority
  * deadline


### 3. PATCH /tasks/{task_id}/complete

📌 Логика:

* Устанавливает `completed = true`
* Добавляет `completed_at`
* Если уже выполнена → ошибка

### 4. DELETE /tasks/{task_id}

* Удаляет задачу
* Возвращает сообщение

### 5. GET /tasks/search

Параметры:

* `keyword`
* `priority`
* `date_from`
* `date_to`

📌 Поиск:

* В title и description
* По диапазону дат


### 6. GET /tasks/stats

Возвращает:

* общее количество
* выполненные
* невыполненные
* просроченные (`deadline < now`)
* распределение по приоритетам
